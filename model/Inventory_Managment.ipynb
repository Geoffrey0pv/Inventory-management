## Calculo de EOQ Y EPQ

```python
DataFrameCharacterization
```

* Economic Order Quantity (EOQ)

```python
import math
def eoq (D, S, h):
  return math.sqrt((2 * D * S) /  (h))
```

* Economic Production Quantity (EPQ)

```python
import math
def epq(D, S, h, p):
    return math.sqrt(   (2 * D * S) / ((1 - (D / p)) * h)   )
```

```python

def calculate_Annual_Demand(sku):
  demanda_promedio = DataFrameCharacterization.loc[DataFrameCharacterization['SKUS'] == sku, "Total Demanda"].iloc[0]
  return demanda_promedio

def fixed_Costs(sku):
  costo_por_ordenar = DataFrameCharacterization.loc[DataFrameCharacterization['SKUS'] == sku, "Costo Ordenar"].iloc[0] # El costo de ordenas un sku por pedido
  return costo_por_ordenar

def holding_Costs(sku):
    costo_por_mantener = DataFrameCharacterization.loc[DataFrameCharacterization['SKUS'] == sku, "Cargo de Mantener una unidad en 90 Semanas"].iloc[0]
    costo_unitario = unit_Cost(sku)
    return costo_por_mantener * costo_unitario # Se multiplica el costo por mantener un sku especifico por su costo unitario

def unit_Cost(sku):
    return DataFrameCharacterization.loc[DataFrameCharacterization['SKUS'] == sku, "Costo unitario "].iloc[0] #Se saca el costo unitario de un sku

def productionRate(sku):
  return DataFrameCharacterization.loc[DataFrameCharacterization['SKUS'] == sku, "Tasa Pdn (unid por semana)"].iloc[0]*90 #La tasa de pnd de una semana por 90 semanas
```

**Paso 1: Simulación del Reabastecimiento.**


Vamos a simular el reabastecimiento de inventario durante 90 semanas utilizando los valores de EOQ (Economic Order Quantity) y EPQ (Economic Production Quantity) para los diferentes SKUs. Se asumirá que el reabastecimiento es automático (lead time = 0), lo que significa que cualquier pedido realizado en una semana estará disponible de inmediato.

```python
def simular_reabastecimiento(df, eoq_epq_dict):
    # Iniciar con el inventario inicial (asumiendo que se proporciona en una columna 'Inv Inicial')
    df['Pedido'] = 0
    df['Recibo'] = 0

    for index, row in df.iterrows():
        # Calcula la demanda de la semana y actualiza el inventario
        if index > 0:  # Asegura que usamos el inventario final de la semana anterior como inicial para la nueva
            df.at[index, 'Inv Final'] = df.at[index - 1, 'Inv Final']

        df.at[index, 'Inv Final'] -= row['Demanda'] + row['Pedido']

        # Revisa si es necesario reabastecer
        if df.at[index, 'Inv Final'] <= 0:
            sku = row['SKU']
            order_quantity = eoq_epq_dict.get(sku, 0)
            df.at[index, 'Pedido'] = order_quantity
            df.at[index, 'Recibo'] = order_quantity
            # Asegurarnos de que el inventario se actualiza considerando el nuevo stock recibido
            df.at[index, 'Inv Final'] += order_quantity


    return df

# Asegurarse de que eoq_epq_dict contiene los valores correctos para cada SKU
eoq_epq_dict = {
    'SKU1': sku1_eoq,  # Ejemplo: Asumir que el EOQ para SKU1 es 1000
    'SKU2': sku2_eoq,
    'SKU3': sku3_eoq,
    'SKU4': sku4_epq,
    'SKU5': sku5_epq,
    'SKU6': sku6_epq,
    'SKU7': sku7_eoq,
    'SKU8': sku8_eoq,
    'SKU9': sku9_eoq,
    'SKU10': sku10_epq,
    'SKU11': sku11_epq,
    'SKU12': sku12_epq,
}

# Llamar a la función reabastecer y asignar el resultado a la variable df si se desea mantener los cambios

Inventory_EPQ_EOQ_Simulation = simular_reabastecimiento(dataInventory1, eoq_epq_dict)
Inventory_EPQ_EOQ_Simulation
```

Para calcular el tamaño del lote usando el método de Period Order Quantity (POQ), necesitamos integrar varios aspectos de tu data, incluyendo la demanda, los costos de mantener y ordenar inventario, y la capacidad de producción. El POQ se basa en encontrar un equilibrio entre estos costos y la capacidad de producción, de manera que minimicemos el costo total y mantengamos un nivel de servicio adecuado.

### Pasos para calcular el POQ:

1. **Determinar el costo de mantener inventario (H)**: Esta información la tienes en la columna "Cargo de Mantener una unidad en un año" en tu primer DataFrame.

2. **Determinar el costo de hacer un pedido (S)**: Esta información se encuentra en la columna "Costo Ordenar" en tu primer DataFrame.

3. **Calcular la demanda promedio (D)**: Puedes utilizar la información de la columna "Total Demanda" del primer DataFrame y dividir por el número de periodos (semanas) si quieres la demanda anual, o directamente usar "Promedio semana".

4. **Determinar la tasa de producción (P)**: Esto se refiere a cuántas unidades puedes producir en una semana, disponible en "Tasa Pdn (unid por semana)".

### Fórmula POQ:

La cantidad de pedido óptimo para el POQ se calcula mediante la fórmula:
$ Q = \sqrt{\frac{2DS}{H}} \times \sqrt{\frac{P}{P-D}} $
donde:
- \( Q \) es la cantidad óptima de pedido.
- \( D \) es la demanda promedio por semana.
- \( S \) es el costo de hacer un pedido.
- \( H \) es el costo de mantener una unidad en inventario por semana.
- \( P \) es la tasa de producción por semana.

### Simulación de Reabastecimiento:

Una vez calculado el tamaño del lote para cada SKU, puedes simular el reabastecimiento en tu segundo DataFrame de la siguiente manera:

- **Inicio de Semana**: El inventario inicial de la semana.
- **Demanda Semanal**: Lo restas del inventario inicial.
- **Final de Semana**: Si el inventario después de la demanda cae por debajo de un punto de reorden, haces un pedido de tamaño \( Q \), el cual se recibirá según el tiempo de entrega.
- **Recibo de Pedido**: Añades los pedidos recibidos al inventario final de la semana para obtener el inventario inicial de la siguiente semana.

Este es un ejemplo básico de cómo podrías proceder. Debes ajustar los detalles específicos según tus datos y necesidades operativas, especialmente los tiempos de entrega y los puntos de reorden.

## Heuristic (Silver-meal)

### **Aplicación del Método de Silver-Meal**

El método de Silver-Meal es una heurística de programación de lotes que busca minimizar el costo total por unidad de tiempo al decidir cuándo y cuánto reordenar. Es especialmente útil para demandas que no son constantes (irregulares).

#### Paso a Paso del Método de Silver-Meal

1. **Inicialización**: Empieza con el primer periodo como el inicio del lote.
2. **Cálculo de Costos**: Calcula el costo total por unidad de tiempo para cada periodo adicional agregado al lote.
3. **Decisión**: Si agregar otro periodo al lote reduce el costo promedio por unidad de tiempo, inclúyelo en el lote. Si no, termina el lote y comienza otro.

```python
import pandas as pd

def silver_meal(df, initial_inventory=0):
    df['Inv_On_Hand'] = 0
    df['Inv_Carry_fwd'] = 0
    df['Make'] = 0
    df['Make_Qty'] = 0
    df['Accum_Make'] = 0
    df['Man_Cost'] = 0
    df['Hld_Cost'] = 0
    df['Z_Costs'] = 0
    df['SM_Calc'] = 0

    P = 0
    N = 0
    M = 0
    RT = 0  # running total of Z_Costs per manufacturing period
    RT2 = 0  # running total of accum_make qty per manufacturing period

    while P + N <= df.iloc[:, 0].max() - 1:

        if df.iloc[N + P, 0] == 1:
            df.iloc[N + P, 4] = int(initial_inventory)  # Inv_On_Hand
        else:
            df.iloc[N + P, 4] = df.iloc[N + P - 1, 5]

        if df.iloc[N + P, 4] - df.iloc[N + P, 1] > 0:
            df.iloc[N + P, 5] = df.iloc[N + P, 4] - df.iloc[N + P, 1]
        else:
            df.iloc[N + P, 5] = 0  # Inventory Carry Forward

        if df.iloc[N + P, 1] > df.iloc[N + P, 4]:
            df.iloc[N + P, 7] = df.iloc[N + P, 1] - df.iloc[N + P, 4]
        else:
            0  # Make_Qty

        RT2 = RT2 + df.iloc[N + P, 7]  # Accum_make
        df.iloc[N + P, 8] = RT2

        if df.iloc[N + P, 8] == df.iloc[N + P, 7] > 0:
            df.iloc[N + P, 6] = 1
            M = 1
        else:
            0
            M = 0  # Make

        df.iloc[N + P, 9] = df.iloc[N + P, 3] * M  # Man_Cost
        df.iloc[N + P, 10] = df.iloc[N + P, 2] * df.iloc[N + P, 7] * N + df.iloc[N + P,
                                                                              2] * df.iloc[
                                                                                   N + P, 5]  # Hold Costs
        # print(N)
        df.iloc[N + P, 11] = df.iloc[N + P, 9] + df.iloc[N + P, 10]  # Z_Costs Sum of Costs

        RT = RT + df.iloc[N + P, 11]

        df.iloc[N + P, 12] = round(RT / (N + 1), 2)  # Average Costs per period for SM calculation

        # df['N_Track'] = N

        # print(df)

        # loop after first run and compare after the second run

        if df.iloc[N + P, 12] > df.iloc[N + P - 1, 12] and N != 0 and df.iloc[N + P, 7] > 0:
            P = N + P
            N = 0
            RT = 0
            RT2 = 0

        else:
            N += 1

    total = 0
    Tot_Man_Costs = 0
    Tot_Hold_Costs = 0

    for i in range(df.shape[0]):
        total += df.iloc[i, 11]
        Tot_Man_Costs += df.iloc[i, 9]
        Tot_Hold_Costs += df.iloc[i, 10]

    print(df)
    print("Total ", total, "Total Manufacturing Costs ", Tot_Man_Costs, "Total Holding Costs ", Tot_Hold_Costs)

    export_to_excel = df.to_excel("Lot Sizes with Irregular Demand.xlsx", index=False, header=True)

# Llamar a la función con tu DataFrame
# silver_meal(df)
```

El algoritmo de Wagner-Whitin es un método utilizado en la gestión de inventarios y la planificación de la producción para determinar el programa de producción óptimo y los niveles de inventario asociados con el fin de minimizar los costos totales.

Se considera una extensión natural y dinamica del problema del modelo EOQ, dónde la demanda es conocida pero presenta una variabilidad durante el periodo de planeación

```python
import numpy as np

def wagner_whitin(demanda, costo_configuracion, costo_mantenimiento):
    n = len(demanda)
    C = np.zeros((n, n))
    setup_costs = np.zeros(n)

    for i in range(n):
        setup_costs[i] = costo_configuracion[i]
        for j in range(i, n):
            for k in range(i, j+1):
                C[i, j] += (k - i + 1) * costo_mantenimiento[k]
            C[i, j] += setup_costs[i]

    # Calcular los costos mínimos utilizando programación dinámica
    min_costs = np.zeros(n)
    min_costs[0] = C[0, 0]
    for j in range(1, n):
        min_costs[j] = C[0, j]
        for i in range(1, j+1):
            cost = min_costs[i-1] + C[i, j]
            if cost < min_costs[j]:
                min_costs[j] = cost

    return min_costs

def calcular_costos_reabastecimiento(DataFrameInventory, DataFrameCharacterization):
    skus_manufacturados = DataFrameCharacterization.loc[DataFrameCharacterization['Tipo'] == 'Manufacturado', 'SKUS'].unique()
    demanda_list = []
    costo_configuracion_list = []
    costo_mantenimiento_list = []

    for sku in skus_manufacturados:
        demanda_sku = DataFrameCharacterization.loc[DataFrameCharacterization['SKUS'] == sku, 'Total Demanda'].values[0]  # Seleccionar el primer valor del array
        costo_configuracion_sku = DataFrameCharacterization.loc[DataFrameCharacterization['SKUS'] == sku, 'Costo Ordenar'].values[0]  # Seleccionar el primer valor del array
        costo_mantenimiento_sku = DataFrameCharacterization.loc[DataFrameCharacterization['SKUS'] == sku, 'Cargo de Mantener una unidad en 90 Semanas'].values[0]  # Seleccionar el primer valor del array

        demanda_list.append(demanda_sku)
        costo_configuracion_list.append(costo_configuracion_sku)
        costo_mantenimiento_list.append(costo_mantenimiento_sku)

    # Calcular los costos óptimos utilizando el algoritmo de Wagner-Whitin
    costos_optimos = wagner_whitin(demanda_list, costo_configuracion_list, costo_mantenimiento_list)

    # Agregar la columna de costos óptimos al DataFrameCharacterization
    DataFrameCharacterization.loc[DataFrameCharacterization['Tipo'] == 'Manufacturado', 'Costos Reabastecimiento'] = costos_optimos
    return DataFrameCharacterization



DataFrameCharacterizationWWW = calcular_costos_reabastecimiento(dataInventoryWW, DataFrameCharacterizationWW)

DataFrameCharacterizationWWW
```

```python
import pandas as pd


# Definimos la función qué realiza el algoritmo Wagner Within

def wagner_within(df, hold_cost, setup_cost):
    Hold_Cost = hold_cost
    Setup_Cost = setup_cost

    df_shape = df.shape
    Max_Row = df_shape[0]
    Max_Col = df_shape[1]

    df.iloc[0, 1] = Setup_Cost  # row then col

    row_index = 0
    col_index = 2
    col_index_start = 1
    Min_cost = 0
    RT = 0

    while row_index < Max_Row:
        while col_index < Max_Col:
            if row_index == 0:
                Min_cost = Setup_Cost
                RT = RT + df.iloc[Max_Row-2, col_index] * (df.iloc[Max_Row - 1, col_index] - col_index_start) * Hold_Cost
                df.iloc[row_index, col_index] = RT + Min_cost
                col_index += 1
            else:
                Min_cost = df.iloc[0:Max_Row - 2, col_index_start-1].min()
                RT = RT + df.iloc[Max_Row - 2, col_index] * (df.iloc[Max_Row - 1, col_index] - col_index_start) * Hold_Cost
                df.iloc[row_index, col_index] = RT + Min_cost
                col_index += 1
        row_index += 1
        col_index_start += 1
        col_index = col_index_start
        Min_cost = 0
        RT = Setup_Cost

    min_idx = Max_Col-1
    task = ""
    replenishment = [0 for i in range(df.shape[0] - 2)]

    while min_idx > 0:
        min_idx_old = min_idx
        min_value = df.iloc[0:Max_Row - 2, min_idx_old].min()
        min_idx = df.index[df.iloc[0:Max_Row - 2, min_idx_old].idxmin(axis=0)]
        make_qty = str(df.iloc[Max_Row-2, min_idx+1:min_idx_old + 1].sum())
        task = ("In period " + str(min_idx + 1) + " make quantity " + make_qty + "   ") + task
        replenishment[min_idx] = make_qty

    order = 0
    min_K = None

    for j in range(df.shape[0] - 2):
        if min_K is None or min_K > df.iloc[j, 90]:
            min_K = df.iloc[j, 90]
            order = df.iloc[j, 0]

    print(f"El mínimo costo es de {min_K} utilizando {order}")

    return replenishment, min_K, order
```
